```python
from sklearn.preprocessing import LabelEncoder
lr=LabelEncoder()
columns=['school','sex','address','famsize','Pstatus','Mjob','Fjob','reason','guardian','schoolsup','famsup','paid','activities','nursery','higher','internet','romantic']
df[columns]=df[columns].apply(lr.fit_transform)
```

```python
df
```

```python
x=df.drop(['G3'],axis=1)
y=df['G3']
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=100)
rf.fit(train_x, train_y)
```

```python
rf.score(test_x,test_y) # .8251
                        # .8178
```
